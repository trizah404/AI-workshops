```python
import pandas as pd
# Load Titanic dataset
df = pd.read_csv("Titanic-Dataset.csv")

print(df.info())

#1
fare = df[['Fare']]
print(fare.head())

#2
class_age = df[['Pclass', 'Age']]
print(class_age.head())

#3
survived_gender = df[['Survived', 'Sex']]
print(survived_gender.head())


```

```python
fare_gt_100 = df[df['Fare'] > 100]
print("Fare > 100:\n", fare_gt_100, "\n")

first_class = df[df['Pclass'] == 1]
print("First Class Passengers:\n", first_class, "\n")

female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("Female Passengers Under 18:\n", female_under_18, "\n")

```

```python
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print("Embarked at C or S:\n", embarked_c_or_s, "\n")

first_second_class = df[df['Pclass'].isin([1, 2])]
print("First or Second Class Passengers:\n", first_second_class)
```

```python
#Exploratory Data Analysis Practice Exercise - 1
#handling missing values 
# Fill missing Age values with the median age before calculation
df['Age'] = df['Age'].fillna(df['Age'].median())

# new column: fare divided by age
df['fare_per_year'] = df['Fare'] / df['Age']

# filterinf passengers whose fare_per_year is greater than 5
high_fare_age = df[df['fare_per_year'] > 5]

# soerrting in descending order of fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

#  only Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

print(result)


```

```python
#Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
#class?

# creating new column fare divided by class
df['fare_per_class'] = df['Fare'] / df['Pclass']

# filtering only adult male passengers
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

# sorting in descending order of fare_per_class
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# keeping Name, Age, and fare_per_class columns
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]


```

```python
#3.3 Exploratory Data Analysis with Group-by Method Practice Exercise

total_fare = df['Fare'].sum()

fare_first = df[df['Pclass'] == 1]['Fare'].sum()
fare_second = df[df['Pclass'] == 2]['Fare'].sum()
fare_third = df[df['Pclass'] == 3]['Fare'].sum()

fare_totals = [fare_first, fare_second, fare_third]

fare_percent = [f / total_fare * 100 for f in fare_totals]

print(fare_percent)
```

```python

# fillingg missing ages with median to avoid issues
df['Age'] = df['Age'].fillna(df['Age'].median())

# creating age group column
df['age_group'] = pd.cut(df['Age'], bins=[0, 17, 64, 200], labels=['child', 'adult', 'senior'])

total_passengers = len(df)

age_group_counts = df['age_group'].value_counts()

age_group_percent = (age_group_counts / total_passengers) * 100

print(age_group_percent)
```
